```python
#import pandas library
import pandas as pd
```

```python
#read csv files
localTemp = pd.read_csv('Mecca_temp.csv', encoding='utf-8')
globalTemp = pd.read_csv('global_temp.csv', encoding='utf-8')
```

<font size = "2">As it can be seen table, the null value from 3 to 17. So for local data, I'll drop records from 1 to 17. Then I'll deal with data from year 1861 until 2013.
Also, global data will be used from 1861 to 2013.</font>

<font size = "4"><b>Cleaning local data</b>

<font size = "2"> As can be seen previously, the dataset of local temperatures has 15 records missing. So, I'll drop those records from data to calculate moving average.</font>

```python
localTemp = localTemp[18:]
```

```python
localTemp.head()
```

```python
globalTemp = globalTemp[111:]
```

```python
globalTemp.head()
```

```python
localTmpeMovAvg = localTemp['avg_temp'].rolling(10).mean()
globalTmpeMovAvg = globalTemp['avg_temp'].rolling(10).mean()
```

```python
localTmpeMovAvg.describe()
```

```python
globalTmpeMovAvg.describe()
```

<font size = "5"><b>Line chart</b></font>

```python
#import tool of visualization
import matplotlib.pyplot as plt
```

```python
#Local and global line chart
plt.figure(figsize=(8, 4))
plt.plot(localTemp['year'],localTmpeMovAvg,label='Mecca')
plt.plot(globalTemp['year'],globalTmpeMovAvg,label='Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperatures")
plt.title("Mecca and Global Average Temperatures")
plt.grid(True)
plt.show()
```

<font size = "5"><b>observations</b></font> 

* It can be seen descriptive of Mecca temprture, the highest average of temperature is 26.800 and the lowest is 24.935. However, the highest average temperature of global is 9.594 and the lowest temperature is 8.595.

* As it can be seen in the line chart, Mecca is hotter than global. Also, the line chart shows Mecca is always hotter over time. 

* The temperatures of Mecca is ups and downs until 1995 then Temperatures are incresed to highest. However, the global temperatures ups and downs until 1980 then it rise to highest.

* From line chart, both Mecca and global are getting hotter over last hundreds years.

## Dataset Overview

PISA2012 dataset is a very rich dataset with 636 columns and 4885490 rows. PISA 2012 is a survey of students' skills and knowledge as they approach the end of compulsory education. This survey is about how well-prepared they are for life beyond school.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
pisa_df = pd.read_csv('pisa_clean.csv')
```

## (Distribution of Scores)

The three histograms show the normal distribution of scores in Math, Reading, and Science. It can be seen, the red line shows the median of scores which is less than 500 points.

```python
# Distribution of Math, Reading, and Science scores

dataList = ['Math Score', 'Reading Score', 'Science Score']
plt.figure(figsize = [30, 8]) 

# histogram
c=1
for i in dataList:
    plt.subplot(1, len(dataList), c) 
    bins = np.arange(0,pisa_df[i].max()+20,20)
    sb.histplot(pisa_df[i], bins = bins, color=sb.color_palette()[0], kde=True)
    plt.axvline(pisa_df[i].median(),  color='r')
    plt.title('Distrebution of '+ i , fontsize=20)
    plt.xlabel(i+ ' (points)')
    plt.ylabel('Count');
    c=c+1
```

## (Relationship between Scores and Gender)

In this part, the box plots show the relationship between scores in Math, Reading, and Science with genders. All scores are nearly the same except the reading score is highest for females.

```python
#relationship plot for math, reading and science with gender

plt.figure(figsize = [8, 14]) 
plt.subplot(3, 1, 1 )
sb.boxplot(data=pisa_df, y='Gender', x='Math Score',)
plt.xlabel('Math Score (points)')
plt.title('Box plot showing the correlation between Math Score with Gender',fontsize=15);




```

```python
plt.figure(figsize = [8, 14]) 
plt.subplot(3, 1, 2 )
sb.boxplot(data=pisa_df, y='Gender', x='Reading Score',)
plt.xlabel('Reading Score (points)')
plt.title('Box plot showing the correlation between Reading Score with Gender',fontsize=15);
```

```python
plt.figure(figsize = [8, 14]) 
plt.subplot(3, 1, 3 )
sb.boxplot(data=pisa_df, y='Gender', x='Science Score',)
plt.xlabel('Reading Score (points)')
plt.title('Box plot showing the correlation between Science Score with Gender',fontsize=15);
```

## (Correlation between scores, genders, and parents' education levels)

The three box plots illustrate the correlation between subjects' scores, genders, and parents' education levels. It can be seen, the scores of all genders are nearly the same but students who are parents have master's or doctoral certificates have high scores. 


```python
#relationship plot for math, reading and science with gender and parent education level

dataList = ['Math Score', 'Reading Score', 'Science Score']
plt.figure(figsize = [8, 30]) 
plt.subplot(3, 1, 2) 
sb.boxplot(data=pisa_df, hue='Parents_Highest_Education',y='Gender' , x='Math Score', palette=sb.color_palette("Set2") )
plt.title('Box plot showing the correlation between Math Score with gender and parent education level',fontsize=15)
plt.xlabel('Math Score (points)')
plt.legend(loc = 6, bbox_to_anchor = (1.0, 0.5));
```

```python
plt.figure(figsize = [8, 30]) 
plt.subplot(3, 1, 2) 
sb.boxplot(data=pisa_df, hue='Parents_Highest_Education',y='Gender' , x='Reading Score', palette=sb.color_palette("Set2") )
plt.title('Box plot showing the correlation between Reading Score with gender and parent education level',fontsize=15)
plt.xlabel('Reading Score (points)')
plt.legend(loc = 6, bbox_to_anchor = (1.0, 0.5));
```

```python
plt.figure(figsize = [8, 30]) 
plt.subplot(3, 1, 3) 
sb.boxplot(data=pisa_df, hue='Parents_Highest_Education',y='Gender' , x='Science Score', palette=sb.color_palette("Set2") )
plt.title('Box plot showing the correlation betweenScience Score with gender and parent education level',fontsize=15)
plt.xlabel('Science Score (points)')
plt.legend(loc = 6, bbox_to_anchor = (1.0, 0.5));
```
